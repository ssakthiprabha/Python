### **6.1 Number of Companies per Sector**

```python
# Creating a custom figure size of 15 X 7 inches
figure = plt.figure(figsize=[15, 7])

# Using magic of pandas's bar function
excel_data['Sector'].value_counts().plot.bar(color='navy')

# Changing x-ticks label size to 12 and rotating to 90 degrees
plt.xticks(rotation=90, size=12)

# Changing y-ticks value using an array of 19 value with step size of 2
plt.yticks(ticks=np.arange(0, 500, 100), size=12)

# Labelling x-axis with a custom label and size of 14
plt.xlabel(xlabel='Sector', size=14)

# Labelling y-axis with a custom label and size of 14
plt.ylabel(ylabel='Number of Companies', size=14)

# Labelling title with a custom label and size of 16
plt.title(label='Number of Companies per Sector', size=16)


# Dispaly the output by rendering visual on the screen
plt.show()
```

```python
# Creating a custom figure size of 10 X 10 inches
figure = plt.figure(figsize=[10, 10])

# Exploding space based on the number of categories
space = np.ones(11)/10

# Using magic of pandas pie() function
excel_data['Sector'].value_counts().plot(kind='pie', fontsize=14,
                                        autopct='%3.1f%%', wedgeprops=dict(width=0.15),
                                        shadow=True, startangle=160, cmap='inferno', legend=True)

plt.ylabel(ylabel='Sector', size=14)
plt.title(label='Donut Plot showing the proportion of each Category value', size=16)
plt.show()
```

### **6.3 Sales, Profits, Assets and Market Value Analysis**

- The value 1E10 is equal to 10000000000

```python
PairPlotsData = excel_data[['Sales','Profits','Assets','Market Value','Country/Territory']]


plt.rc('xtick', labelsize=12)
plt.rc('ytick', labelsize=12)
plt.rc('axes', labelsize=14)

#Set Style of Pair Plots
sns.set_style('darkgrid')
sns.color_palette("crest", as_cmap=True)

#sns.pairplot(PairPlotsData,  plot_kws={'color':'black'})
# Create a Pair Plots
sns.pairplot(PairPlotsData, diag_kind='kde', kind='reg',
             diag_kws={'color': 'black', 'shade':True, 'lw': 2}
            )
plt.tight_layout();
```

### **6.4 Profits, Sales and  Assets Analysis of Top 100 Companies | 3D Graph **

- The value 1E10 is equal to 10000000000

```python
Scatter3D_data = excel_data.head(1000) 

threedee = plt.figure(figsize=(15, 15)).gca(projection='3d')
threedee.scatter3D(Scatter3D_data['Profits'].index, Scatter3D_data['Sales'], Scatter3D_data['Assets'], s=100)

threedee.set_xlabel('Profits')
threedee.set_ylabel('Sales')
threedee.set_zlabel('Assets')

threedee.set_xlim3d(-17900000000, 88200000000)
threedee.set_ylim3d(200000000, 524000000000)
threedee.set_zlim3d(500000000, 4322500000000)

plt.show()
```

### **6.5 Profits, Sales Analysis of Bottom 100 Companies **

```python
#Analysis of Bottom 100 Data

JointPlotBottomData = excel_data[['Sales','Profits','Country/Territory','Industry']].tail(100) 

sns.set_style('darkgrid')
sns.jointplot(x=JointPlotBottomData.Sales, y=JointPlotBottomData.Profits,height=8,space=0.1,ratio=3,kind='reg');
```
